# ☁️ FLUX.1-dev Cloud Image Generation

Welcome to the **FLUX.1-dev** generation notebook! 
Normally, this model requires a giant graphics card and 25GB of disk space. But here on **Google Colab**, you can use Hugging Face's Serverless Interface to generate images instantly!

### Step 1: Get your Hugging Face Token
1. You MUST Accept the FLUX.1-dev license here: [Black Forest Labs FLUX.1-dev](https://huggingface.co/black-forest-labs/FLUX.1-dev)
2. Get your free token from: [Hugging Face Settings](https://huggingface.co/settings/tokens)
3. Paste it in the cell below and run it (Shift + Enter).

```python
import requests
from IPython.display import Image, display

# ⚠️ REPLACE 'YOUR_TOKEN_HERE' WITH YOUR ACTUAL HUGGING FACE TOKEN!
HF_TOKEN = "YOUR_TOKEN_HERE"

API_URL = "https://api-inference.huggingface.co/models/black-forest-labs/FLUX.1-dev"
headers = {"Authorization": f"Bearer {HF_TOKEN}"}

print("Libraries loaded. Ready to generate!")
```

---
### Step 2: Generate the Art 🎨
Modify the `prompt` variable below to whatever you want to draw. Then run the cell!

```python
# WRITE YOUR PROMPT HERE:
prompt = "A highly detailed cyberpunk cat wearing neon sunglasses, standing in the rainy streets of Tokyo at night, cinematic lighting, 4k"

print(f"[*] Sending prompt to FLUX.1-dev: '{prompt}'")
print("[*] Generating... (this usually takes 5-10 seconds depending on server load)")

payload = {"inputs": prompt}
response = requests.post(API_URL, headers=headers, json=payload)

if response.status_code == 200:
    # Save the output to the Colab environment
    output_filename = "flux_colab_image.jpg"
    with open(output_filename, "wb") as file:
        file.write(response.content)
    
    print("\n✅ Success! Here is your generated image:")
    # Display the image right inside the Notebook!
    display(Image(filename=output_filename))
    
    print("\n(You can right-click the image to save it, or download it from the left menu)")
else:
    print(f"\n❌ Error {response.status_code}: {response.text}")
    if response.status_code == 401:
        print("\n👉 DID YOU FORGET TO PUT YOUR TOKEN IN THE TOP CELL?")
    elif response.status_code == 403:
        print("\n👉 YOU MUST GO TO HUGGINGFACE AND CLICK 'AGREE AND ACCESS REPOSITORY' TO ACCEPT THE FLUX LICENSE FIRST!")
    elif "loading" in response.text.lower():
        print("\n⏳ The server is just waking up the giant model. Run this cell again in 30 seconds!")
```
