### 🎨 Generating Art with Code (No Visual Nodes!)
We will now write the exact equivalent of a ComfyUI visual workspace using Python variables.

```python
# 3. Import all ComfyUI Nodes as if they were Python Functions
from comfyscript.nodes import *

# We use 'Workflow()' context to tell ComfyScript we are building a graph
with Workflow():
    # 1. Load the Model (Checkpoint)
    # ComfyUI will automatically download this small default model from HuggingFace!
    model, clip, vae = CheckpointLoaderSimple('v1-5-pruned-emaonly.safetensors')
    
    # 2. Positive Prompt (What we want)
    positive = CLIPTextEncode('masterpiece, best quality, a beautiful cyberpunk city in the rain, neon lights, 4k', clip)
    
    # 3. Negative Prompt (What we DO NOT want)
    negative = CLIPTextEncode('ugly, blurry, low resolution, bad anatomy', clip)
    
    # 4. Blank Canvas (Empty Latent Image)
    # (Width=512, Height=512, BatchSize=1)
    latent = EmptyLatentImage(512, 512, 1)
    
    # 5. KSampler (The Engine that denoises the canvas based on prompts)
    # Args: model, seed, steps, cfg, sampler_name, scheduler, positive, negative, latent_image, denoise
    sampled_latent = KSampler(model, 123456789, 20, 8.0, 'euler_ancestral', 'normal', positive, negative, latent, 1.0)
    
    # 6. VAE Decode (Convert the math back into pixels)
    image = VAEDecode(sampled_latent, vae)
    
    # 7. Save and Preview Image
    SaveImage(image, 'My_First_ComfyScript_Output')

print("Workflow built and executed! Check the completely generated image on the left folder menu.")
```
